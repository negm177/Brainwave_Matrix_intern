```python
# üè¨ Total Sales by Branch
branch_sales = df.groupby('Branch')['Total'].sum().reset_index()
branch_sales = branch_sales.sort_values(by='Total', ascending=False)
branch_sales
```

```python
# üìä Visualize total sales per branch
plt.figure(figsize=(8,5))
sns.barplot(x='Branch', y='Total', data=branch_sales)
plt.title('Total Sales per Branch')
plt.xlabel('Branch')
plt.ylabel('Total Sales')
plt.show()
```

```python
# üì¶ Quantity Sold by Product Line
products_sales = df.groupby('Product line')['Quantity'].sum().reset_index()
products_sales = products_sales.sort_values(by='Quantity', ascending=False)
products_sales
```

```python
# üìä Visualize quantity sold per product line
plt.figure(figsize=(12,5))
sns.barplot(x='Product line', y='Quantity', data=products_sales)
plt.title('Quantity Sold by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Quantity')
plt.xticks(rotation=45)
plt.show()
```

```python
# üë• Total Sales by Gender
gender_sales = df.groupby('Gender')['Total'].sum().reset_index()
gender_sales = gender_sales.sort_values(by='Total', ascending=False)
gender_sales
```

```python
# ü•ß Visualize sales distribution by gender
plt.figure(figsize=(6,6))
plt.pie(gender_sales['Total'], labels=gender_sales['Gender'], autopct='%1.1f%%', startangle=140)
plt.title('Total Sales by Gender')
plt.axis('equal')
plt.show()
```

```python
# üìä Average Sales by Gender
gender_avg = df.groupby('Gender')['Total'].mean().reset_index()
gender_avg
```

```python
# üìä Count of each gender
gender_count = df['Gender'].value_counts().reset_index()
gender_count.columns = ['Gender', 'Count']
gender_count
```

```python
# üõçÔ∏è Product Line Preference by Gender
product_gender = df.groupby(['Gender','Product line'])['Total'].sum().reset_index()
product_gender = product_gender.sort_values(by='Total', ascending=False)
product_gender
```

```python
# üìä Visualize gender-wise product sales
plt.figure(figsize=(12,5))
sns.barplot(x='Product line', y='Total', hue='Gender', data=product_gender)
plt.title('Product Sales by Gender')
plt.xlabel('Product Line')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```
